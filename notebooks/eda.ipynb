```python
import pandas as pd
import numpy as np

df_raw = pd.read_csv('../data/Placement_Data_Full_Class.csv')

df_clean = df_raw.copy()
df_clean = df_clean.drop(columns=['sl_no', 'salary'])
df_clean = df_clean.dropna()

df_clean['status'] = df_clean['status'].map({'Placed': 1, 'Not Placed': 0})

df_clean.to_csv('../data/cleaned_placement.csv', index=False)
```
