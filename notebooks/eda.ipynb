```python
import pandas as pd
import numpy as np

df_raw = pd.read_csv('../data/2018_Financial_Data.csv')

keep_cols = [
    'returnOnEquity', 
    'Debt to Equity', 
    'PE ratio', 
    'Revenue Growth', 
    'operatingProfitMargin',
    'Class'
]

df_clean = df_raw[keep_cols].copy()

df_clean = df_clean.dropna()

q_low = df_clean.quantile(0.01)
q_hi = df_clean.quantile(0.99)
df_clean = df_clean[~((df_clean < q_low) | (df_clean > q_hi)).any(axis=1)]

df_clean.to_csv('../data/cleaned_fundamentals.csv', index=False)
```
